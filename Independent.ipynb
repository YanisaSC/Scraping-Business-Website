```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import datetime
```

```python
baseurl = 'https://www.independent.co.uk/asia'
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:93.0) Gecko/20100101 Firefox/93.0'}
```

```python
r = requests.get('https://www.independent.co.uk/news/business?CMP=ILC-refresh')

soup = BeautifulSoup(r.content, 'lxml')

newlist = soup.find_all('h2', class_='sc-gATJxy iaYZKe')
    
newlink = []
    
for item in newlist:
    for link in item.find_all('a', href=True): 
        newlink.append(baseurl + link['href'])
newlink = list(set(newlink))        
print(len(newlink))
```

```python
newlink.remove('https://www.independent.co.uk/asia/author/phil-thornton')
newlink.remove('https://www.independent.co.uk/asia/author/james-moore')
newlink.remove('https://www.independent.co.uk/asia/author/andy-martin')
newlink.remove('https://www.independent.co.uk/asia/author/chris-blackhurst')
newlink.remove('https://www.independent.co.uk/asia/author/anna-isaac')
newlink.remove('https://www.independent.co.uk/asia/author/martin-friel')
```

```python
print(len(newlink))
```

```python
testlink ='https://www.independent.co.uk/news/business/amazon-drivers-workers-rights-b1937660.html'

r = requests.get(testlink, headers = headers)
    
soup = BeautifulSoup(r.content, 'lxml')

title = soup.find('h1', class_='sc-DGwzl gyVzcL').text.strip()

content =  soup.select('p', style ='sc-fWSMSg brpdtc main-wrapper ') 

publishdatetime = soup.find('div', class_= 'sc-bgSLBM kCekDr').text.strip()

print(title)
```

```python
#testlink ='https://www.independent.co.uk/news/business/amazon-drivers-workers-rights-b1937660.html'
newlist = []
for link in newlink:
    r = requests.get(link, headers = headers)
    soup = BeautifulSoup(r.content, 'lxml')
    title=soup.find_all('div',class_='sc-gWnPEt zignS')
    for i in title:
        for j in i.find_all('h1'):
            title=j.text
    publishdatetime = soup.find('amp-timeago').text.strip()
    
    retrievedatetime = (datetime.datetime.now()).strftime("%x")
    source_url = link

    content = soup.find_all('div', class_='sc-fWSMSg brpdtc main-wrapper')
    content2=""
    for i in content:
        for a in i.find_all("p"):
            text = a.text
            content2 += text+" "
    news = {
            'title': title,
            'publish datetime': publishdatetime,
            'retrievedatetime': retrievedatetime,
            'source_url': source_url,
            'content': content2,
            }
    newlist.append(news)
```
