```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import datetime
```

```python
baseurl = 'https://www.businesstoday.co/'

headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/94.0.4606.71 Safari/537.36'}

r = requests.get('https://www.businesstoday.co/category/business/')

soup = BeautifulSoup(r.content, 'lxml')

newslist = soup.find_all('div',class_='td-module-thumb')

newslinks = []

for item in newslist:
    for link in item.find_all('a', href = True):
        newslinks.append(link['href'])
newslinks = list(dict.fromkeys(newslinks))        
print(newslinks)       
print(len(newslist))
    
```

```python
#testlink = 'https://www.businesstoday.co/business/property/14/09/2021/76040/'
def getnews(newslinks):
    newslist = []
    for link in newslinks:

        r = requests.get(link, headers = headers)
        soup = BeautifulSoup(r.content, 'lxml')

        title = soup.find('h1', class_='tdb-title-text').text.strip()
        publishdatetime = soup.find('time', class_='entry-date updated td-module-date')['datetime']

        retrievedatetime = (datetime.date.today())

        source_url = link

        content = soup.find_all('div', class_='tdb-block-inner td-fix-index')
        content2 = ''
        for x in content:
            for i in x.find_all('p'):
                content2 += i.text

        news = {
            'title': title,
            'publish datetime': publishdatetime,
            'retrievedatetime': retrievedatetime,
            'source_url': source_url,
            'content': content2,
            }
        newslist.append(news)
    return newslist
data=getnews(newslinks)
    
```
