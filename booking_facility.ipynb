* สร้าง function สำหรับจัดการข้อมูลสิ่งอำนวยความสะดวกยอดนิยม (โดยทำการ select ทีละตัว และ strip ตัวที่เป็น /n หรือ อักขระพิเศษทิ้ง)

```python
def scrap_facility(select_facility):
    important_facility=[]
    for i in select_facility:
        str1 = i.get_text()
        new_str = str1.strip()
        important_facility.append(new_str)
    return important_facility
```

* สร้าง function สำหรับจัดการข้อมูลชื่อโรงแรม จัดการตัดสิ่งที่ไม่ใช่ชื่อโรงแรมออก แล้วคืนค่าเป็นชื่อโรงแรม 


```python
def scrap_name(select_title):
    name_soup=select_title
    name=name_soup[0].get_text()
    sub=re.sub(r'โรงแรม/n','',name)
    sub=re.sub(r'/n','',sub)
    text=sub.strip()
    return(text)
```

* สร้าง funtion สำหรับจัดการข้อมูลชื่อหัวคะแนน  มีการตัดตัวเลขออก โดยให้ค่าออกมาเป็นชื่อหัวคะแนน (เช่น คะแนนพนักงาน คะแนนสิ่งอำนวยความสะดวก คะแนนความสะอาด คะแนนความสะดวกสบาย และคะแนนไวไฟฟรี เป็นต้น)


```python
def scrap_score_bar_title(select_score_bar,i):
    c=select_score_bar
    text=c[i].get_text().strip()
    title=re.sub(r'[0-9.]','',text)
    return title
```

* สร้าง funtion สำหรับจัดการข้อมูลคะแนน ใช้คำสั่ง re.findall หาข้อมูลตัวเลขและ
กำหนดค่าข้อมูลเป็น float โดยที่ให้ค่าออกมาเป็นคะแนน เช่น 8.9


```python
def scrap_score_bar_score(select_score_bar,i):
    c=select_score_bar
    text=c[i].get_text().strip()
    score = re.findall(r'[0-9.]',text)
    score=''.join(score)
    score=float(score)
    return score
```

* ทำการเก็บ Name (ชื่อโรงแรม), Facility (สิ่งอำนวยความสะดวก), และ Scorebar_dict (คะแนนจากผู้เข้าพัก) เข้าไปใน list 
 * แต่เนื่องจากเราเก็บข้อมูลออกมาแล้วพบว่าได้ข้อมูลซ้ำออกมาถึง 3 ชุด เราจึงทำการหาร 3


```python
for i in run_list:
    r = requests.get(i, headers = headers)
    soup = BeautifulSoup(r.content, 'lxml')
    try:
        name=scrap_name(soup.select('.hp__hotel-name')).strip()
        facility=scrap_facility(soup.select('.important_facility'))
        scorebar_dict={}
        for i in range ((len(soup.select('.c-score-bar')))//3):
            title=scrap_score_bar_title(soup.select('.c-score-bar'),i).strip()
            score=scrap_score_bar_score(soup.select('.c-score-bar'),i)
            scorebar_dict[title]=score
        hotel = {
            'ชื่อ': name,
            'สิ่งอำนวยความสะดวก':facility,
            'คะแนน':scorebar_dict
            }
        hotel_list.append(hotel)            
    except Exception as e:
        None
```

เนื่องจากเราเก็บข้อมูลออกมาแล้วพบว่าได้ facility ซ้ำออกมาถึง 2 ชุด จึงสร้าง funtion เพื่อเลือกข้อมูลเพียงชุดเดียว

```python
def half_list(data):
    amount=len(data)
    data=data[:amount//2]
    return data     
```

* ทำการลบประโยคส่วนเกินด้านหน้าใน Column “name” ออก เพื่อให้ประโยคเหมือนกัน โดยกำหนดชื่อว่า clean_name (ให้เหลือแต่ชื่อโรงแรม)

```python
def clean_name(text):
    sub=re.sub(r'บ้านพัก\n','',str(text))
    sub=re.sub(r'ลอดจ์\n','',str(sub))
    sub=re.sub(r'เบดแอนด์เบรกฟาสต์\n','',str(sub))
    sub=re.sub(r'อพาร์ตเมนต์\n','',str(sub))
    sub=re.sub(r'อินน์\n','',str(sub))
    sub=re.sub(r'รีสอร์ท\n','',str(sub))
    sub=re.sub(r'เซอร์วิสอพาร์ตเมนต์\n','',str(sub))
    sub=re.sub(r'วิลลา\n','',str(sub))
    sub=re.sub(r'โฮมสเตย์\n','',str(sub))
    sub=re.sub(r'บ้านพักตากอากาศ\n','',str(sub))
    sub=re.sub(r'โฮสเทล\n','',str(sub))
    sub=re.sub(r'โรงแรม\n','',str(sub))
    #sub=re.sub(r'.+[\n]+','',str(text))
    return sub
```

```python
export['name']=export['name'].apply(clean_name)
```
