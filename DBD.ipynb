```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import datetime
```

```python
baseurl = 'https://www.dbd.go.th/'
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:93.0) Gecko/20100101 Firefox/93.0'}
```

```python
for page in range(2,51):
    r = requests.get('https://www.dbd.go.th/more_news.php?cid=252&page={}')

    soup = BeautifulSoup(r.content, 'lxml')

    newlist = soup.find_all('li', class_='col-md-12')
    
    newlink = []
    
    for item in newlist:
        for link in item.find_all('a', href=True): 
            newlink.append(baseurl + link['href'])
        
print(len(newlink))
```

```python
#testlink ='https://www.dbd.go.th/news_view.php?nid=469421039'
def getnews(newlink):
    newlist = []
    for link in newlink:
        r = requests.get(link, headers = headers)
        soup = BeautifulSoup(r.content, 'html.parser')
        title = soup.find('h4', class_='article-title font30').text.strip()
        publishdatetime = soup.find('ul', class_= 'article-meta').text.strip()
        retrievedatetime = (datetime.datetime.now()).strftime("%x")
        source_url = link
        content =  soup.find_all('div', class_ ='article-body')    
        content2 = ""
        for i in content:
            span_select=i.find_all("span")
            div_select=i.find_all("div")
            if len(span_select) >= len(div_select):
                for a in span_select:
                    text = a.text
                    if text != title:
                        content2 += text+" "
            else:
                for a in div_select:
                    text = a.text
                    if text != title:
                        content2 += text+" "
        news = {'title': title,
                'publish datetime': publishdatetime,
                'retrievedatetime': retrievedatetime,
                'source_url': source_url,
                'content': content2,
                }
        newlist.append(news)
    return newlist
print(getnews(newlink))
```

```python
df = pd.DataFrame(getnews(newlink))
```
